## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/Task (2) Dataset.csv')
df.head()
```

```python
df.describe()
```

### Identify and analyze patterns in the data.

```python
df['City'].value_counts()
```

```python
df2= df.groupby('City')['JamsCount'].sum()
print(df2)
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
plt.figure(figsize=(10, 6))
df['City'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('muted'))
plt.show()
```

```python
sns.histplot(df['JamsCount'], kde=True, bins=20, color='blue')
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
sns.scatterplot(data=df,x= 'City' , y='JamsCount'  )
```
